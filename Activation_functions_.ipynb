#Activation Functions

1.Sigmoid:

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+np.exp(-x))
x=np.linspace(-10,11)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation function : sigmoid')
plt.show()
```

2.Tanh

```python
import numpy as np
import matplotlib.pyplot as plt
def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,11)
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('Activation function : Tanh')
plt.show()
```

3.reLU

```python
import numpy as np
import matplotlib.pyplot as plt
def reLU(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
      x1.append(i)
  return x1
x=np.linspace(-10,11)
plt.plot(x,reLU(x))
plt.axis('tight')
plt.title('Activation function : reLU')
plt.show()
```

4.softmax

```python
import numpy as np
import matplotlib.pyplot as plt
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x),axis=0)
x=np.linspace(-10,11)
plt.plot(x,softmax(x))
plt.axis('tight')
plt.title('Activation function : softmax')
plt.show()
```
